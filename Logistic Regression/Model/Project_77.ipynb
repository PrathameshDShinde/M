```python
#### Importing Liabraries
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline
```

```python
#### Getting the data
def read_data():
    try:
        data = pd.read_csv("C:/Data Science/Nov 2023/19 Nov 2023/Logistic Regresssion_Project - 7/4. Logistic Regresssion/bank-full.csv")
        return data
    except Exception as e:
        print("Data Path Not Found and Error is in the Read Data Function")
```

```python
#### Calling the read data function
df = read_data()
print(df.head())
```

```python
# #### Preprocessing and feature engineering
df1 = df.copy()
print(df1.head())
print(df1.info())
print(df1.shape)
```

```python
plt.figure(figsize=(12,4))
ax = sns.barplot(data=df1,x='job',y='age',errorbar=None,estimator='count',hue='marital',palette='viridis')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Martial Statuswise Job")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Martial Statuswise Job")
plt.show()
```

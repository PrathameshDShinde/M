```python
#### Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report

#### Getting the data
def read_data():
    try:
        data = pd.read_csv("C:/Data Science/Nov 2023/19 Nov 2023/Decision Tree Project - 8/Decision Tree  - Assignment -4/iphone_purchase_records.csv")
        return data
    except Exception as e:
        print("Data Path Not Found and Error is in the Read Data Function")

#### Calling the read data function
df = read_data()
# print(df.head())

#### Preprocessing and Feature Engineering
df1 = df.copy()
# print(df.head())
# print(df.info())
# print(df.shape)
```

```python
df1
```

```python
plt.figure(figsize=(12,4))
type_value_counts = df['Gender'].value_counts().head()
colors = sns.color_palette('coolwarm',len(type_value_counts))
plt.pie(type_value_counts,labels=type_value_counts.index,
       autopct='%1.2f%%',startangle=140,colors=colors,wedgeprops=dict(width=0.4))
plt.title('Percentage of Gender')
plt.tight_layout()
plt.savefig('Percentage of Gender')
plt.show()
```

### Inferance - In this data 51% are Female and 49% are Male.

```python
plt.figure(figsize=(10,4))
purchase_iphone_value_counts = df1['Purchase Iphone'].value_counts().head()
colors = sns.color_palette('viridis',len(purchase_iphone_value_counts))
plt.pie(purchase_iphone_value_counts,labels=purchase_iphone_value_counts.index,
       autopct='%1.2f%%',startangle=140,colors=colors,wedgeprops=None)
plt.title('Freuency of Purchasing Iphone')
plt.tight_layout()
plt.savefig('Freuency of Purchasing Iphone')
plt.show()
```

### Inferance - In this data Iphon purchasing rate in 36% and 64% are not purchasing Iphone. 

```python
plt.figure(figsize=(8,4.5))
ax = sns.barplot(data = df1,x = 'Gender',y = 'Age',errorbar=None,hue='Purchase Iphone',palette='viridis',estimator='count' )
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Which gender purchase more Iphone ?')
plt.tight_layout()
plt.savefig('Which gender purchase more Iphone')
plt.show()
```

### Inferance - Female are purchasing more Iphone as compared to Male.

```python
plt.figure(figsize=(8,4.5))
ax = sns.barplot(data = df1,x = 'Gender',y = 'Salary',errorbar=None,hue='Purchase Iphone',palette='viridis',estimator='count' )
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Salary wise purchasing Iphone')
plt.tight_layout()
plt.savefig('Salary wise purchasing Iphone')
plt.show()
```

```python
plt.figure(figsize=(8,4.5))
ax = sns.barplot(data = df1,x = 'Gender',y = 'Salary',errorbar=None,hue='Purchase Iphone',palette='viridis',estimator='sum' )
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Total Salary wise purchasing Iphone')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4.5))
ax = sns.scatterplot(data = df1,x = 'Age',y = 'Salary',hue='Purchase Iphone',palette='viridis' )
plt.title('Corelation Between Age and Salary')
plt.tight_layout()
plt.savefig('Corelation Between Age and Salary')
plt.show()
```
